```python
from operator import truediv
#1)

# Señal cuadrada (escalón)

import matplotlib.pyplot as plt

def señal_cuadrada(t, periodo=1, amplitud=1):

    if (t % periodo) < (periodo / 2):
        return amplitud
    else:
        return -amplitud



t_val = [i * 0.01 for i in range(1000)]
señal_cuadrada_val = [señal_cuadrada(t, periodo=2, amplitud=1) for t in t_val]


plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(t_val, señal_cuadrada_val, label='Señal Cuadrada')
plt.title("Señal Cuadrada")
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()

# Señal triangular

def señal_triangular(t, periodo=1, amplitud=1):
    t_normalizado = t % periodo
    fase = t_normalizado / (periodo / 2)
    if fase <= 1:
        return amplitud * fase
    else:
        return amplitud * (2 - fase)

t_val = [i * 0.01 for i in range(1000)]
señal_triangular_val = [señal_triangular(t, periodo=2, amplitud=1) for t in t_val]

plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 2)
plt.plot(t_val, señal_triangular_val, label='Señal Triangular', color='orange')
plt.title("Señal Triangular")
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()






#2)
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal

num_muestras = 1000  # Número de muestras
frecuencia = 5  # Frecuencia de las señales en Hz
frec_muestreo = 1000  # Frecuencia de muestreo

# Señal cuadrada (escalón)

señal_cuadrada = signal.square(2 * np.pi * frecuencia * np.arange(num_muestras) / frec_muestreo)

plt.figure(figsize=(12, 8))

plt.subplot(4, 1, 2)
plt.plot(señal_cuadrada)
plt.title('Señal Cuadrada')
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()

# Señal delta de dirac

señal_dirac = np.zeros(num_muestras)
señal_dirac[0] = 1

plt.figure(figsize=(12, 8))

plt.subplot(4, 1, 4)
plt.plot(señal_dirac, color='red')
plt.title('Delta de Dirac')
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()

# Señal triangular

señal_triangular = signal.sawtooth(2 * np.pi * frecuencia * np.arange(num_muestras) / frec_muestreo, 0.5)

plt.figure(figsize=(12, 8))

plt.subplot(4, 1, 3)
plt.plot(señal_triangular)
plt.title('Señal Triangular')
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()

# Señal Senoidal

señal_senoidal = np.sin(2 * np.pi * frecuencia * np.arange(num_muestras) / frec_muestreo)

plt.figure(figsize=(12, 8))

plt.subplot(4, 1, 1)
plt.plot(señal_senoidal, color='orange')
plt.title('Señal Senoidal')
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()








#3)

fs = 500  # Frecuencia de muestreo
t = np.linspace(0, 1, fs, endpoint=False)  # Tiempo entre 0 y 1 segundo, con una frecuencia de muestreo fs

# Señal cuadrada (escalón)

frecuencia_cuadrada = 5  # Frecuencia de la señal cuadrada en Hz
señal_cuadrada = signal.square(2 * np.pi * frecuencia_cuadrada * t)

plt.figure(figsize=(12, 8))

plt.subplot(4, 1, 2)
plt.plot(t, señal_cuadrada, color='cyan')
plt.title('Señal Cuadrada')
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()



# Señal delta de dirac


señal_dirac = np.zeros_like(t)
señal_dirac[0] = 1

plt.figure(figsize=(12, 8))

plt.subplot(4, 1, 4)
plt.plot(t, señal_dirac, color='red')
plt.title('Delta de Dirac')
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()

# Señal triangular

frecuencia_triangular = 5  # Frecuencia de la señal triangular en Hz
señal_triangular = signal.sawtooth(2 * np.pi * frecuencia_triangular * t, 0.5)  # parámetro 0.5 para que sea triangular

plt.figure(figsize=(12, 8))

plt.subplot(4, 1, 3)
plt.plot(t, señal_triangular, color='yellow')
plt.title('Señal Triangular')
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()


# Señal Senoidal

frecuencia_seno = 5  # Frecuencia de la señal senoidal en Hz
señal_senoidal = np.sin(2 * np.pi * frecuencia_seno * t)

plt.figure(figsize=(12, 8))

plt.subplot(4, 1, 1)
plt.plot(t, señal_senoidal, color='green')
plt.title('Señal Senoidal')
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()





#4)

import math

def generar_senal_cuadrada(frecuencia, num_muestras, fs):
    senal = []
    for i in range(num_muestras):
        t = i / fs
        valor = 1 if math.sin(2 * math.pi * frecuencia * t) >= 0 else -1
        senal.append(valor)
    return senal

def generar_senal_dirac(num_muestras):
    senal = [0] * num_muestras
    senal[0] = 1
    return senal

def generar_senal_triangular(frecuencia, num_muestras, fs):
    senal = []
    for i in range(num_muestras):
        t = i / fs
        valor = 2 * (2 * (frecuencia * t - math.floor(frecuencia * t + 0.5)))
        senal.append(valor)
    return senal

def generar_senal_senoidal(frecuencia, num_muestras, fs):
    senal = []
    for i in range(num_muestras):
        t = i / fs
        valor = math.sin(2 * math.pi * frecuencia * t)
        senal.append(valor)
    return senal

def encontrar_maximos_minimos(señal):
    valor_maximo = max(señal)
    valor_minimo = min(señal)

    # Encontrar todos los índices del valor máximo y mínimo
    indices_maximos = [i for i, valor in enumerate(señal) if valor == valor_maximo]
    indices_minimos = [i for i, valor in enumerate(señal) if valor == valor_minimo]

    return indices_maximos, valor_maximo, indices_minimos, valor_minimo

# Parámetros de la señal
num_muestras = 1000
frecuencia = 5
fs = 1000

# Generación de señales
senal_cuadrada = generar_senal_cuadrada(frecuencia, num_muestras, fs)
senal_dirac = generar_senal_dirac(num_muestras)
senal_triangular = generar_senal_triangular(frecuencia, num_muestras, fs)
senal_senoidal = generar_senal_senoidal(frecuencia, num_muestras, fs)

# Máximos y mínimos
max_min_cuadrada = encontrar_maximos_minimos(senal_cuadrada)
max_min_dirac = encontrar_maximos_minimos(senal_dirac)
max_min_triangular = encontrar_maximos_minimos(senal_triangular)
max_min_senoidal = encontrar_maximos_minimos(senal_senoidal)


def imprimir_resultados(nombre, max_min):
    print(f"Señal {nombre}:")
    print(f"Máximo en índices {max_min[0]}, Valor: {max_min[1]:.4f}")
    print(f"Mínimo en índices {max_min[2]}, Valor: {max_min[3]:.4f}\n")

# Mostrar resultados
imprimir_resultados("Cuadrada", max_min_cuadrada)
imprimir_resultados("Delta de Dirac", max_min_dirac)
imprimir_resultados("Triangular", max_min_triangular)
imprimir_resultados("Senoidal", max_min_senoidal)








#5)

def encontrar_maximos_minimos(señal):
    valor_maximo = max(señal)
    valor_minimo = min(señal)

    # Índices del valor máximo y mínimo
    indices_maximos = [i for i, valor in enumerate(señal) if valor == valor_maximo]
    indices_minimos = [i for i, valor in enumerate(señal) if valor == valor_minimo]

    return indices_maximos, valor_maximo, indices_minimos, valor_minimo


import numpy as np
from scipy import signal

num_muestras = 1000
frecuencia = 5
fs = 1000

# Señales
señal_cuadrada = signal.square(2 * np.pi * frecuencia * np.arange(num_muestras) / fs).tolist()
señal_dirac = np.zeros(num_muestras).tolist()
señal_dirac[0] = 1
señal_triangular = signal.sawtooth(2 * np.pi * frecuencia * np.arange(num_muestras) / fs, 0.5).tolist()
señal_senoidal = np.sin(2 * np.pi * frecuencia * np.arange(num_muestras) / fs).tolist()

# Máximos y mínimos
max_min_cuadrada = encontrar_maximos_minimos(señal_cuadrada)
max_min_dirac = encontrar_maximos_minimos(señal_dirac)
max_min_triangular = encontrar_maximos_minimos(señal_triangular)
max_min_senoidal = encontrar_maximos_minimos(señal_senoidal)


def imprimir_resultados(nombre, max_min):
    print(f"Señal {nombre}:")
    print(f"Máximo en índices {max_min[0]}, Valor: {max_min[1]:.4f}")
    print(f"Mínimo en índices {max_min[2]}, Valor: {max_min[3]:.4f}\n")

# Mostrar resultados
imprimir_resultados("Cuadrada", max_min_cuadrada)
imprimir_resultados("Delta de Dirac", max_min_dirac)
imprimir_resultados("Triangular", max_min_triangular)
imprimir_resultados("Senoidal", max_min_senoidal)















```
